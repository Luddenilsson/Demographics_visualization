```python
scb.set_query(region="Riket", kön="totalt", ålder=["20-24 år", "25-29 år", "30-34 år", "35-39 år"], 
              tabellinehåll= ['Medelinkomst, tkr', 'Medianinkomst, tkr', 'Totalsumma, mnkr', 'Antal personer'],
              år= ['1991', '1992',  '1993',  '1994',  '1995',  '1996',  '1997',  '1998',  '1999',  
                   '2000',  '2001',  '2002',  '2003',  '2004',  '2005',  '2006',  '2007',  '2008',
                   '2009',  '2010',  '2011',  '2012',  '2013',  '2014',  '2015',  '2016',  '2017',
                   '2018',  '2019',  '2020',  '2021','2022'])
```

```python
scb.get_query()
```

```python
income = scb.get_data()
```

```python
income["data"]
```

```python
# Create a DataFrame from the list
df = pd.DataFrame(income["data"])

# Extract values into separate columns
df[['Medelinkomst, tkr', 'Medianinkomst, tkr', 'Totalsumma, mnkr', 'Antal personer']] = pd.DataFrame(df['values'].tolist(), index=df.index)

# Convert columns to appropriate data types
df['Medelinkomst, tkr'] = df['Medelinkomst, tkr'].astype(float)
df['Medianinkomst, tkr'] = df['Medianinkomst, tkr'].astype(float)
df['Totalsumma, mnkr'] = df['Totalsumma, mnkr'].astype(float)
df['Antal personer'] = df['Antal personer'].astype(int)

# Extract age and year from 'key' column
df[['Age', 'Year']] = pd.DataFrame(df['key'].apply(lambda x: [x[2], x[3]]).tolist(), index=df.index)

# Set 'Age' and 'Year' as the index
df.set_index(['Age', 'Year'], inplace=True)

# Drop unnecessary columns
df.drop(columns=['key', 'values'], inplace=True)

# Display the resulting DataFrame
df
```

```python
df.to_csv("income_91to22.csv")
```

```python
scb.go_down("PR", "PR0101", "PR0101A", "KPItotM")
scb.info()
```

```python
scb.get_variables()
```

```python
scb.set_query(tabellinnehåll='KPI, fastställda tal', månad=scb.get_variables()["månad"])
```

```python
inflation = scb.get_data()['data']
```

```python
inflation
```

```python
# Create a DataFrame from the list
df = pd.DataFrame(inflation)

# Extract values into separate columns
df['KPI-tal'] = pd.DataFrame(df['values'].tolist(), index=df.index)

# Convert columns to appropriate data types
df['KPI-tal'] = df['KPI-tal'].astype(float)

# Extract age and year from 'key' column
df['Year'] = pd.DataFrame(df['key'].apply(lambda x: [x[0]]).tolist(), index=df.index)

# Set 'Age' and 'Year' as the index
df['Year'] = pd.to_datetime(df['Year'], format='%YM%m')
df.set_index('Year', inplace=True)

# Drop unnecessary columns
df.drop(columns=['key', 'values'], inplace=True)

# Display the resulting DataFrame
df = df.resample("Y", kind="period").mean()
```

```python
df.to_csv("KPI_1980.csv")
```
