```python
# Plotting
sns.set(style="white")
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
i = 0
# Use a loop to create a separate line for each age group
for age_group, group_data in income.groupby('Age'):
    median_KPI = KPI.loc[2022]['KPI-tal']*group_data['Medianinkomst, tkr']/KPI.loc[1991:2022]['KPI-tal']
    df = median_KPI.reset_index()
    df.rename(columns={df.keys()[2]: 'value'}, inplace=True)
    ax = axes[int(i/2)][i%2]
    sns.lineplot(data=df, x='Year', y='value', label=f'', ax=ax)
    ax2 = ax.twinx()
    sns.lineplot(data=fertility.loc[fertility['År'] > 1990], x='År', y=age_group + ' ', color='red', label=f'', ax=ax2)
    i += 1
    ax.set_ylabel('Yearly Income')
    ax2.set_ylabel('Births per 1000 women')
    ax.set_xlabel(None)
    plt.title(age_group + ' year olds', fontsize=14)

plt.suptitle('Yearly Income(kkr in 2022 value) compared to fertility', fontsize=20)
#plt.legend()
plt.tight_layout()
sns.despine(top=True,
    right=True,
    left=True,
    bottom=True,
    offset=None,
    trim=True)
plt.savefig('Graphs/income_ferility.png', transparent=True)
plt.show()
```

```python
# Create a FacetGrid with a separate plot for each age group
g = sns.FacetGrid(unemployment, col="Age", col_wrap=1, height=6)
g.map(plt.plot, "Year", "ALÖS", color='blue', label='ALÖS')
#g.map(plt.plot, "Year", "SYS", color='green', label='SYS')

# Set titles and labels
g.set_axis_labels("Year", "Values")
g.set_titles(col_template="{col_name}")
g.tight_layout()

# Show the legend for each subplot
g.add_legend()

# Iterate through each subplot and plot fertility data
for age_group, ax in zip(unemployment['Age'].unique(), g.axes):
    ax2 = ax.twinx()
    #fertility_data = fertility[fertility['Age'] == age_group]
    if age_group == '25-34':
        ax2.plot(fertility['År'].loc[fertility['År'] > 2000], (fertility['25-29 '].loc[fertility['År'] > 2000] + fertility['30-34 '].loc[fertility['År'] > 2000])/1000
                 , color='red', label=f'Fertility {age_group} 20-24')
    elif age_group == '35-44':
        ax2.plot(fertility['År'].loc[fertility['År'] > 2000], (fertility['35-39 '].loc[fertility['År'] > 2000] + fertility['40-44 '].loc[fertility['År'] > 2000])/1000
                 , color='red', label=f'Fertility {age_group} 20-24')
    else:
        ax2.plot(fertility['År'].loc[fertility['År'] > 2000], fertility[age_group + ' '].loc[fertility['År'] > 2000]/1000, color='red', label=f'Fertility {age_group} 20-24')
    ax2.set_ylabel(f'Fertility {age_group}')

# Show the legend for the right y-axis
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.2))

# Show the plot
plt.show()
```
